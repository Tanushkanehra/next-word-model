```python
#CONVERTING THE TEXT DATA INTO SEPERATE SENTANCES
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
path='technology.txt'
text = open(path).read()
text2=sent_tokenize(text)
text2
```

```python
#SEPERATING THE WORDS FROM THE SENTANCES
ntoken.fit_on_texts(text)
listwords=text_to_word_sequence(text)
main=list(listwords)
main
```

```python
#ASSIGNING THE NUMERICAL VALUE TO EACH WORD
ntoken.fit_on_texts(text2)
seq=ntoken.texts_to_sequences(text2)
```

```python
#NUMERICAL VALUE GIVEN TO THE SINGLE WORD
ntoken.word_index
```

```python

x=[]
y=[]
total_word_drop=0
for i in seq:
    if len(i)>1:
        for index in range(1,len(i)):
            x.append(i[:index])
            y.append(i[index])
    else:
        total_word_drop+=1
print(total_word_drop)
```

```python
#MAKING THE LSTM MODEL2
model = Sequential()
model.add(Embedding(vocab_size,14))
model.add(LSTM(150))
model.add(Dense(vocab_size, activation='softmax'))
print(model.summary())
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=50)
```

```python
def make_prediction2(text, n_words):
    for i in range(n_words):
        text_tokenize = ntoken.texts_to_sequences([text])
        text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
        prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))
        prediction = str(vocab_array[prediction - 1])
        print(vocab_array[np.argsort(model.predict(text_padded)) - 1].ravel()[:-3])
        text += " " + prediction
    return text
```
